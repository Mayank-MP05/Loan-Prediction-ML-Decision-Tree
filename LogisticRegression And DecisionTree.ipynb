```python
df = pd.read_csv('/kaggle/input/cleanloan/Loan-Clean-Data.csv',index_col=0)
df.head()
```

```python
X = df.drop(labels=['Loan_Status_Y'],axis=1,inplace=False)
Y = df['Loan_Status_Y']
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(X,Y,test_size=0.17,random_state=13)
```

### Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(x_train,y_train)
```

```python
log
```

```python
log.score(x_train,y_train)
```

#### 81% Accuracy - Given by model

```python
# lets test it out on real data
from sklearn.metrics import accuracy_score as ac
pred=log.predict(x_test)
ac(y_test,pred)
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier as dtc
dtc = dtc()
```

```python
dtc.fit(x_train,y_train)
```

```python
dtc.score(x_train,y_train)
```

###### almost same accuracy as of Logistic Regression

```python
treePred = dtc.predict(x_test)
ac(y_test,treePred)
```

### 72% Real Tested accuracy

```python
## Confusion Matrix for Decision Tree Classiifier 
from sklearn import metrics
metrics.confusion_matrix(y_test,pred)
```
